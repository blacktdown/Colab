```
import matplotlib.pyplot as plt
import numpy as np
```

```
#기본차트
plt.title('Plot1')
plt.plot([1, 3, 4, 8, 10, 17])
plt.show()
```

```
#x축 설정
plt.title('Plot2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 8, 10, 17])
plt.show()
```

```
#차트 스타일
plt.title('Plot3')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 8, 10, 17], 'ro--')
plt.grid(True)
plt.show()
```

```
#멀티 라인차트
vec = np.arange(0, 5, 0.2)

plt.title('Plot4')
plt.plot(vec, vec, 'r--')
plt.plot(vec, vec**2, 'b--')
plt.plot(vec, vec**3, 'g--')
plt.grid(True)
plt.show
```
