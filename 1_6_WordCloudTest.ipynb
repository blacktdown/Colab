```
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```
#드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')
```

```
#데이터 불러오기
txt = open('/content/drive/MyDrive/Colab Notebooks/file/wordcloud_data.txt').read()
txt
```

```
#WordCloud 출력
wc = WordCloud(width = 1200,
               height = 1200,
               background_color = 'white',
               font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf').generate(txt)

plt.imshow(wc, interpolation = 'bilinear')
plt.show()
```

```
#한글 형태소 분석기 생성
from konlpy.tag import Okt
okt = Okt()
```

```
#기본 형태소 분석
res1 = okt.pos('나는 어제 학원에서 파이썬을 공부 했습니다.')
res1

```

```
#명사만 추출
noun_txt = okt.nouns(txt)
#리스트를 문자열로 변환
str_noun_txt = " ".join(noun_txt)
```

```
#명사만 WordCloud 단어 시각화
wc = WordCloud(width = 1200,
               height = 1200,
               background_color = 'white',
               font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf').generate(str_noun_txt)

plt.imshow(wc, interpolation='bilinear')
plt.show()
```
